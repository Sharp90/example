```python
import pandas as pd
```

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid
```

### 1. covid 데이터에서 100 case 대비 사망률(Deaths / 100 Cases)이 가장 높은 국가는?

```python
covid_1 = covid.sort_values(by = 'Deaths / 100 Cases', ascending = False)
covid_1
```

```python
covid_1.iloc[0, 0]
```

### 2. covid 데이터에서 신규 확진자가 없는 나라 중 WHO Region이 'Europe'를 모두 출력하면?

```python
covid_2 = covid[covid['New cases'] == 0]
covid_2[covid_2['WHO Region'] == 'Europe'].iloc[:, 0]
```
